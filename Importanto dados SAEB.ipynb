Nos testes aplicados na quarta e oitava séries (quinto e nono anos) do ensino fundamental e terceiro ano do ensino médio, os estudantes respondem a questões de língua portuguesa, com foco em leitura, e matemática, com foco na resolução de problemas. No questionário socioeconômico, os estudantes fornecem informações sobre fatores de contexto que podem estar associados ao desempenho.

Professores e diretores das turmas e escolas avaliadas também respondem a questionários que coletam dados demográficos, perfil profissional e de condições de trabalho.

A Prova Brasil é aplicada em nível nacional a cada dois anos em todas as escolas públicas urbanas e rurais que tenham no mínimo 20 estudantes matriculados nas séries avaliadas. As escolas particulares também podem se cadastrar para participar da prova de forma voluntária.

Infelizmente a prova não é aplicada anualmente o que nos impede de medir a evolução do ensino de forma contínua ano após ano, mas ainda assim a base traz ricos insumos através dos questionários socioeconômicos.

Segundo site qedu.org.br/brasil/aprendizado apenas 34% dos alunos apresentaram aprendizado adequado em língua portuguesa no 9º ano.
Já na disciplina de matemática o percentual foi de 15% - assustador!
Os índices do 5º ano são um pouco melhores que no 9º, mas ainda assim preocupantes: 56% em língua portuguesa e 44% em matemática.

Por estes motivos o foco deste trabalho será nos indices de aprendizagem do 9º ano nas matérias de língua portuguesa e matemática.

##### Meu objetivo é explorar as respostas dos questionários respondidos pelos alunos, diretores e professores e classificar se a escola terá indice de aprendizagem adequado em lingua portuguesa e em matemática nas oitavas séries (nono ano) das escolas brasileiras. 

Adicionalmente pretendo entender quais são as características que fazem cada escola atigir ou não o nível adequado de aprendizado em cada matéria.

O aprendizado adequado foi definindo utilizando o que especialista em Educação do QEDU usam na divulgação dos resultados, conforme links abaixo:

Nível de aprendizado brasileiro:

https://academia.qedu.org.br/prova-brasil/aprendizado-adequado/

Regra de aprendizado adequado:

https://academia.qedu.org.br/prova-brasil/454-2/

Matemática 9º ano: > 300 ou Níveis 5 a 9

Lingua Portuguesa 9º ano: > 275 ou Níveis 4 a 8

#### Base questionário Diretor

```python
#diretor
df_diretor = pd.read_csv('gs://bucket-tcc-fia-mba/SAEB/2017/TS_DIRETOR_2017.csv', sep=',', encoding='utf-8', index_col = 'ID_ESCOLA')
print(df_diretor.shape)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_diretor.head(10))
```

Mantém apenas escolas com questionário preenchido e remove colunas que não correnspodem ao questionário.

```python
df_diretor = df_diretor[df_diretor.IN_PREENCHIMENTO_QUESTIONARIO==1]
df_diretor.drop(columns=['ID_UF', 'ID_MUNICIPIO', 'ID_DEPENDENCIA_ADM', 'ID_LOCALIZACAO', 'IN_PREENCHIMENTO_QUESTIONARIO', 'ID_PROVA_BRASIL'], inplace=True)
print(df_diretor.shape)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_diretor.head(10))
```

Renomeia colunas para identificá-las como pertencentes ao questionário do Diretor.

```python
df_diretor.columns = ['DIRETOR_' + x for i, x in enumerate(df_diretor.columns, 1)]
df_diretor.head(3)
```

#### Base questionário Professor

```python
#professor
df_prof = pd.read_csv('gs://bucket-tcc-fia-mba/SAEB/2017/TS_PROFESSOR_2017.csv', sep=',', encoding='utf-8')
print(df_prof.shape)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof.head(10))
```

Mantém apenas professores que lecionam na nona série e que tenham preenchido o questionário.

```python
df_prof = df_prof[(df_prof.IN_PREENCHIMENTO_QUESTIONARIO==1) & (df_prof.ID_SERIE==9)]
df_prof.drop(columns=['ID_UF', 'ID_MUNICIPIO', 'ID_DEPENDENCIA_ADM', 'ID_LOCALIZACAO', 'ID_TURMA', 'CO_PROFESSOR', 'ID_SERIE', 'IN_PREENCHIMENTO_QUESTIONARIO', 'ID_PROVA_BRASIL'], inplace=True)
print(df_prof.shape)

#filtrar qm tem IN_PREENCHIMENTO_QUESTIONARIO = 1 e ID_SERIE = 9
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof.head(10))
```

Como o objetivo é avaliar a aprendizagem das matérias de lingua portuguesa e matemática, selecionarei apenas os questionários respondidos pelos professores dessas máterias.

```python
df_prof.groupby('TX_RESP_Q105')['TX_RESP_Q105'].count()
```

A = Língua Portuguesa

B = Matemática

C = Mais de uma disciplina.


Separa base de questionários respondidos por professores de lingua portuguesa.

```python
df_prof_lp = df_prof[(df_prof.TX_RESP_Q105=='A')]
display(df_prof_lp.shape)
display(df_prof_lp.head())
```

Separa base de questionários respondidos por professores de matemática.

```python
df_prof_mt = df_prof[(df_prof.TX_RESP_Q105=='B')]
display(df_prof_mt.shape)
display(df_prof_mt.head())
```

Existe mais de um professor da mesma matéria por escola, para manter o indice por escola vou selecionar aquele que preencheu mais itens do questionário e descartar as demais informações.

Lingua Portuguesa

```python
df_prof_lp['qtd_preenchido'] = df_prof_lp.apply(lambda x: x.isnull().sum(), axis = 1 )
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_lp.head(10))
```

```python
df_prof_lp.sort_values(by=['ID_ESCOLA', 'qtd_preenchido'], ascending=[True,True], inplace=True)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_lp.head(20))
```

```python
df_prof_lp_final = df_prof_lp.drop_duplicates(subset=['ID_ESCOLA'])
df_prof_lp_final.set_index('ID_ESCOLA', inplace=True)
df_prof_lp_final.drop(columns='qtd_preenchido', inplace=True)
df_prof_lp_final.shape
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_lp_final.head(10))
```

```python
df_prof_lp_final.columns = ['PROF_LP_' + x for i, x in enumerate(df_prof_lp_final.columns, 1)]
df_prof_lp_final.head(3)
```

Matemática

```python
df_prof_mt['qtd_preenchido'] = df_prof_mt.apply(lambda x: x.isnull().sum(), axis = 1 )
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_mt.head(10))
```

```python
df_prof_mt.sort_values(by=['ID_ESCOLA', 'qtd_preenchido'], ascending=[True,True], inplace=True)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_mt.head(20))
```

```python
df_prof_mt_final = df_prof_mt.drop_duplicates(subset=['ID_ESCOLA'])
df_prof_mt_final.set_index('ID_ESCOLA', inplace=True)
df_prof_mt_final.drop(columns='qtd_preenchido', inplace=True)
df_prof_mt_final.shape
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_mt_final.head(10))
```

```python
df_prof_mt_final.columns = ['PROF_MT_' + x for i, x in enumerate(df_prof_mt_final.columns, 1)]
df_prof_mt_final.head(3)
```
