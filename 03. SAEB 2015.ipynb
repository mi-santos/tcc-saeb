```python
import pandas as pd
import numpy as np
```

### Importando dados da prova SAEB de 2015

```python
#saeb
df_saeb = pd.read_csv('gs://bucket-tcc-fia-mba/SAEB/2015/TS_ESCOLA_2015.csv', sep=',', encoding='latin-1', index_col = 'ID_ESCOLA')
print(df_saeb.shape)
```

```python
with pd.option_context("display.max_rows", 153, "display.max_columns", 153):
    display(df_saeb.head())
```

Deleta colunas de avaliação de outros periodos de ensino.

```python
df_saeb.drop(columns=['PC_FORMACAO_DOCENTE_INICIAL', 'IN_PREENCHIMENTO_QUESTIONARIO',
                      'NU_MATRICULADOS_CENSO_5EF', 'NU_PRESENTES_5EF', 'TAXA_PARTICIPACAO_5EF', 
                      'NIVEL_0_LP5', 'NIVEL_1_LP5', 'NIVEL_2_LP5', 'NIVEL_3_LP5', 'NIVEL_4_LP5', 'NIVEL_5_LP5', 'NIVEL_6_LP5', 'NIVEL_7_LP5', 'NIVEL_8_LP5', 'NIVEL_9_LP5', 
                      'NIVEL_0_MT5', 'NIVEL_1_MT5', 'NIVEL_2_MT5', 'NIVEL_3_MT5', 'NIVEL_4_MT5', 'NIVEL_5_MT5', 'NIVEL_6_MT5', 'NIVEL_7_MT5', 'NIVEL_8_MT5', 'NIVEL_9_MT5', 'NIVEL_10_MT5',
                      'MEDIA_5EF_LP', 'MEDIA_5EF_MT'], inplace=True)
print(df_saeb.shape)
```

Remove escolas que não possuem nono ano.

```python
df_saeb9 = df_saeb[df_saeb['NU_MATRICULADOS_CENSO_9EF'].notnull()]
df_saeb9.shape
```

```python
with pd.option_context("display.max_rows", 153, "display.max_columns", 153):
    display(df_saeb9.head())
```

#### Base questionário Diretor

```python
#diretor
df_diretor = pd.read_csv('gs://bucket-tcc-fia-mba/SAEB/2015/TS_DIRETOR_2015.csv', sep=',', encoding='utf-8', index_col = 'ID_ESCOLA')
print(df_diretor.shape)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_diretor.head(10))
```

Mantém apenas escolas com questionário preenchido e remove colunas que não correnspodem ao questionário.

```python
df_diretor = df_diretor[df_diretor.IN_PREENCHIMENTO_QUESTIONARIO==1]
df_diretor.drop(columns=['ID_UF', 'ID_MUNICIPIO', 'ID_DEPENDENCIA_ADM', 'ID_LOCALIZACAO', 'IN_PREENCHIMENTO_QUESTIONARIO', 'ID_PROVA_BRASIL'], inplace=True)
print(df_diretor.shape)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_diretor.head(10))
```

Renomeia colunas para identificá-las como pertencentes ao questionário do Diretor.

```python
df_diretor.columns = ['DIRETOR_' + x for i, x in enumerate(df_diretor.columns, 1)]
df_diretor.head(3)
```

#### Base questionário Professor

```python
#professor
df_prof = pd.read_csv('gs://bucket-tcc-fia-mba/SAEB/2015/TS_PROFESSOR_2015.csv', sep=',', encoding='utf-8')
print(df_prof.shape)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof.head(10))
```

Mantém apenas professores que lecionam na nona série e que tenham preenchido o questionário.

```python
df_prof = df_prof[(df_prof.IN_PREENCHIMENTO_QUESTIONARIO==1) & (df_prof.ID_SERIE==9)]
df_prof.drop(columns=['ID_UF', 'ID_MUNICIPIO', 'ID_DEPENDENCIA_ADM', 'ID_LOCALIZACAO', 'ID_TURMA', 'ID_SERIE', 'IN_PREENCHIMENTO_QUESTIONARIO', 'ID_PROVA_BRASIL'], inplace=True)
print(df_prof.shape)

#filtrar qm tem IN_PREENCHIMENTO_QUESTIONARIO = 1 e ID_SERIE = 9
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof.head(10))
```

Como o objetivo é avaliar a aprendizagem das matérias de lingua portuguesa e matemática, selecionarei apenas os questionários respondidos pelos professores dessas máterias.

```python
df_prof.groupby('TX_RESP_Q105')['TX_RESP_Q105'].count()
```

A = Língua Portuguesa

B = Matemática

C = Mais de uma disciplina.


Separa base de questionários respondidos por professores de lingua portuguesa.

```python
df_prof_lp = df_prof[(df_prof.TX_RESP_Q105=='A')]
display(df_prof_lp.shape)
display(df_prof_lp.head())
```

Separa base de questionários respondidos por professores de matemática.

```python
df_prof_mt = df_prof[(df_prof.TX_RESP_Q105=='B')]
display(df_prof_mt.shape)
display(df_prof_mt.head())
```

Existe mais de um professor da mesma matéria por escola, para manter o indice por escola vou selecionar aquele que preencheu mais itens do questionário e descartar as demais informações.

Lingua Portuguesa

```python
df_prof_lp['qtd_preenchido'] = df_prof_lp.apply(lambda x: x.isnull().sum(), axis = 1 )
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_lp.head(10))
```

```python
df_prof_lp.sort_values(by=['ID_ESCOLA', 'qtd_preenchido'], ascending=[True,True], inplace=True)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_lp.head(20))
```

```python
df_prof_lp_final = df_prof_lp.drop_duplicates(subset=['ID_ESCOLA'])
df_prof_lp_final.set_index('ID_ESCOLA', inplace=True)
df_prof_lp_final.drop(columns='qtd_preenchido', inplace=True)
df_prof_lp_final.shape
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_lp_final.head(10))
```

```python
df_prof_lp_final.columns = ['PROF_LP_' + x for i, x in enumerate(df_prof_lp_final.columns, 1)]
df_prof_lp_final.head(3)
```

Matemática

```python
df_prof_mt['qtd_preenchido'] = df_prof_mt.apply(lambda x: x.isnull().sum(), axis = 1 )
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_mt.head(10))
```

```python
df_prof_mt.sort_values(by=['ID_ESCOLA', 'qtd_preenchido'], ascending=[True,True], inplace=True)
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_mt.head(20))
```

```python
df_prof_mt_final = df_prof_mt.drop_duplicates(subset=['ID_ESCOLA'])
df_prof_mt_final.set_index('ID_ESCOLA', inplace=True)
df_prof_mt_final.drop(columns='qtd_preenchido', inplace=True)
df_prof_mt_final.shape
```

```python
with pd.option_context("display.max_rows", 165, "display.max_columns", 165):
    display(df_prof_mt_final.head(10))
```

```python
df_prof_mt_final.columns = ['PROF_MT_' + x for i, x in enumerate(df_prof_mt_final.columns, 1)]
df_prof_mt_final.head(3)
```
